```python
def fixBadZipfile(zipFile):  
    f = open(zipFile, 'r+b')  
    data = f.read()  
    pos = data.find(b'\x50\x4b\x05\x06') # End of central directory signature  
    if (pos > 0):  
        print("Trancating file at location " + str(pos + 22)+ ".")  
        f.seek(pos + 22)   # size of 'ZIP end of central directory record' 
        f.truncate()  
        f.close()  
    else:  
        print("ERRORERROR")
         # raise error, file is truncated
```

```python
def get_els(header):
    body=""
    for i in header.next_siblings:
        is_a = False
        if i.em:
            source = i.em.text
            continue
        for d in i.descendants:
            if d.name == "a":
                url=d.text
                is_a = True
        if not is_a:
            body += i.text
    sample_dict = {
        "title": header.text,
        "source": source,
        "body": body,
        "url": url
    }
    return sample_dict

```
